# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

## With lower number of discharges, there is a greater incidence of excess rate of readmissions (area shaded red)
## With higher number of discharges, there is a greater incidence of lower rates of readmissions (area shaded green) 

To break it down, what the author was trying to say was, the average excess rate of readmissions of the hospitals with less than 350 discharges is higher than the abverage excess rate of readmissions of the hospitals with 800-2500 charges.

To test if it's true, we have to run a one_tailed hypothesis testing:  

**H0: the average excess rate of readmissions of the hospitals with less than 350 discharges is no different than the average excess rate of readmissions of the hospitals with 800-2500 charges.**  
**HA: the average excess rate of readmissions of the hospitals with less than 350 discharges is higher than the average excess rate of readmissions of the hospitals with 800-2500 charges.**


```python
import scipy.stats as stats
dischargelowdf=df[df['Number of Discharges']<=350]
dischargehighdf=df[(df['Number of Discharges']>=800) & (df['Number of Discharges']<=2500)]
print (stats.ttest_ind(dischargelowdf['Excess Readmission Ratio'].values,dischargehighdf['Excess Readmission Ratio'].values,equal_var=False))
```

The P-value is really small (< 0.01), reject the null hypothesis.  
The average excess rate of readmissions of the hospitals with less than 350 discharges is higher than the average excess rate of readmissions of the hospitals with 800-2500 charges.

## 2. Compute and report the observed significance value (or p-value).
Again, since we don't know the population mean of each sample, we use the one_tailed t_test

```python
dischargelowdf2=df[df['Number of Discharges']<=350]
dischargehighdf2=df[df['Number of Discharges']>350]
print (stats.ttest_ind(dischargelowdf2['Excess Readmission Ratio'].values,dischargehighdf2['Excess Readmission Ratio'].values,equal_var=False))
```
